```python
import pandas as pd
df=pd.read_csv("Dataset/expenditure.csv")
```

```python
dfs=[df]

for df in dfs:
    df.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 69'], inplace=True)
    df.set_index('Country Name', inplace=True)
```

```python
df
```

```python
# Only year after 2000, best consistency

df_filtered = df.loc[:, [
    col for col in df.columns
    if col.isdigit() and int(col) > 1999
]]

#I drop the most recent columns, as they are null
df_filtered = df_filtered.drop(columns=["2023", "2024"])
```

```python
df_filtered=df_filtered.dropna()
```

```python
# Melt to transform from wide to long format
df_filtered = df_filtered.reset_index().melt(id_vars=['Country Name'], var_name='Year', value_name='Expenditure')
```
